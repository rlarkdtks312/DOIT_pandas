### 넓은 데이터
- 넓은 데이터란 열 자체가 값을 가지는 데이터를 의미한다. 열 자체가 데이터를 가지면서 옆으로 길게 늘어선 형태가 된다.

### melt 메서드
|메서드 인자|설명|
|:-----|:---|
|id_val|위지를 그대로 유지할 열의 이름을 지정한다.|
|value_vars|행으로 위치를 변경할 열의 이름을 지정한다.|
|var_name|value_vars로 위치를 변경할 열의 이름을 지정한다.|
|value_name|var_name으로 위치를 변결할 열의 데이터를 지정한 열의 이름을 저장한다|

### melt 메서드 사용하기

- 1개의 열만 고정하고 나머지 열을 행으로 바꾸기

```python
import pandas as pd
pew = pd.read_csv('./data/pew.csv')
pew.head()
```

```python
pew.iloc[:, 0:6]
```

- id_vars 인자에 고정할 열을 지정하여 소득 정보 열로 정리되었다.
- 열은 variable 변수로 그 값은 value로 나타난다

```python
pew_long = pd.melt(pew, id_vars='religion')
```

```python
pew_long
```

- variable, value 열의 열 이름 변경하기

```python
pew_long = pd.melt(pew, id_vars='religion', var_name='income', value_name='count')
```

```python
pew_long
```

- 2개 이상의 열을 고정하고 나머지 열을 행으로 바꾸기

```python
billboard = pd.read_csv('./data/billboard.csv')
billboard.iloc[0:5, 0:16]
```

- wkn 열만 피벗하기

```python
billboard_long = pd.melt(billboard, id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
                        var_name='week', value_name='rating')
billboard_long.head()
```
