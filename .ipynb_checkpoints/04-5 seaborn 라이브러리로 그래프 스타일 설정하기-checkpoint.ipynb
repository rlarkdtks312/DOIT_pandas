```python
import seaborn as sns
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.set_style('dark')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.set_style('white')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.set_style('ticks')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```
