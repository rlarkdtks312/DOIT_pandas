### 가상 데이터를 여러 열을 하나로 정리하기 - melt, pivot_table 메서드

```python
import pandas as pd
```

```python
weather = pd.read_csv('./data/weather.csv')
weather.head()
```

```python
weather_melt = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temp')
weather_melt.head()
```

- pivot_table 메서드는 행과 열의 위치를 다시 바꿔 정리한다.
- index 인자에는 위치를 그대로 유지할 열 이름을 지정하고, columns 인자에는 피벗할 열 이름을 지정하고, value 인자에는 새로운 열의 데이터가 될 열 이름을 지정한다.

```python
weather_tidy = weather_melt.pivot_table(
    index = ['id', 'year', 'month', 'day'],
    columns='element',
    values='temp',
    dropna=False
    )
weather_tidy
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat
```
