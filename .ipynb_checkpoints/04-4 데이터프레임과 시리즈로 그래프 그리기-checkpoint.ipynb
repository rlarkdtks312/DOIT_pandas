```python
import pandas as pd
import seaborn as sns
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

#### 데이터 프레임과 시리즈로 그래프 그리기

```python
ax = plt.subplots()
ax = tips['total_bill'].plot.hist()
```

```python
fig, ax = plt.subplots()
ax = tips[['total_bill', 'tip']].plot.hist(alpha=0.5, bins=20, ax=ax)
```

```python
ax = plt.subplots()
ax = tips['tip'].plot.kde()
```

```python
ax = plt.subplots()
ax = plt.scatter(data=tips, x='total_bill', y='tip')
```

```python
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill', y='tip', ax=ax)
```

```python
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill', y='tip', gridsize=10, ax=ax) # gridsize 인자를 통한 육각형의 크기 변경
```

```python
fig, ax = plt.subplots()
ax = tips.plot.box(ax=ax)
```
