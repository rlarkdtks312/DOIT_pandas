### 빌보드 차트의 중복 데이터 처리하기

```python
import pandas as pd
```

```python
billboard = pd.read_csv('./data/billboard.csv')
billboard.head()
```

```python
billboard_long = pd.melt(billboard, id_vars=['year', 'artist', 'track', 'time', 'date.entered'], var_name='week', value_name='rating')
billboard_long.head()
```

```python
# 노래 제목(track)이 Loser인 데이터가 중복 되어있다는 것을 확인 할 수 있다.
# 가수(artist)는 고유한 값이기 때문에 따로 관리하는 것이 데이터의 일관성을 유지하는데 더 도움이 된다.
billboard_long[billboard_long.track == 'Loser']
```

```python
# 중복 데이터를 가지고 있는 열은 year, artist, track, time, date이다. 이 열을 따로 모아 새로운 데이터프레임에 저장한다.
billboard_songs = billboard_long[['year', 'artist', 'track', 'time', 'date.entered']]
billboard_songs
```

```python
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs
```

```python
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs
```

```python
# merge 메서드를 사용해 노래 정보와 주간 순위 데이터를 합친다.
billboard_rating = billboard_long.merge(billboard_songs, on=['year', 'artist', 'track', 'time', 'date.entered'])
```
