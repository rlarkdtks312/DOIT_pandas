### 하나의 열이 여러 의미를 가지고 있는 경우
- ebola 데이터 집합의 Deaths_Guinea는 사망자 수와 나라 이름을 합쳐 만든 이름이다.

#### ebola 데이터 집합 살펴보기

```python
import pandas as pd
ebola = pd.read_csv('./data/country_timeseries.csv')
ebola
```

```python
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
ebola_long
```

### split 메서드로 열 이름 분리하기
- Cases_Guinea와 같이 2개 이상의 의미를 가지고 있는 열 이름은 밑줄(_)을 기준으로 분리할 수 있다.

#### 열 이름 분리하고 데이터프레임에 추가하기

```python
variable_split = ebola_long.variable.str.split('_')
variable_split
```

```python
print(type(variable_split))
```

```python
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)

print(status_values[:5])
print(country_values[:5])
print(status_values[-5:])
print(country_values[-5:])
```

```python
ebola_long['status'] = status_values
ebola_long['country'] = country_values
ebola_long.head()
```

#### concat 메서드로 데이터프레임에 열 추가하기
- concat 메서드를 활용하면 split 메서드로 분리한 데이터를 바로 데이터 프레임에 추가할 수 있다.

```python
variable_split = ebola_long.variable.str.split('_')
variable_split.colums = ['status', 'counrty']
ebola_parsed = pd.concat([ebola_long, variable_split], axis=1)
ebola_parsed
```
